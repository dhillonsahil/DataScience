```python
# Sample data
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]

plt.plot(x1, y1, marker='d', linestyle='-')

# Adding labels and title
plt.xlabel("X-axis Label")  # Set x-axis label
plt.ylabel("Y-axis Label")  # Set y-axis label
plt.title("Graph Title")  # Set title

# Show the plot
plt.show()
```

Other marker options:

    'o'  Circle
    's'  Square
    '^'  Triangle up
    'v'  Triangle down
    'D'  Diamond
    'x'  Cross
    '+'  Plus sign
    'p'  Pentagon
    'h'  Hexagon

Other linestyle options:

    '-'  Solid line (default)
    '--'  Dashed line
    ':'  Dotted line
    '-.'  Dash-dot line
    '' (empty string)  No line, only markers

```python
x= np.random.rand(50)
y= np.random.rand(50)

plt.scatter(x,y)
```

```python
colours = np.random.rand(50)
sizes=np.random.rand(50)*50
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title("Random Data")
plt.scatter(x,y,c=colours,s=sizes,alpha=0.5) # s for diff size of each dot by sizes array , c to change color of each dot by colours array , alpha to show intensity of color
plt.show()
```

```python
plt.scatter(x,y,c=colours,s=sizes,alpha=1) 
```

```python
plt.bar(x,y)
plt.xlabel("Name")
plt.ylabel("Age")
```

```python
# showing a 3d :- 1st make a figure add subplot to tell project then show graph
x= np.random.rand(50)
y= np.random.rand(50)
z= np.random.rand(50)

fig = plt.figure()
axis=fig.add_subplot(projection= '3d')

axis.scatter(x,y,z)
```

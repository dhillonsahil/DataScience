```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
iris_data=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
x_scaled=scaler.fit_transform(iris_data)
```

```python
x_scaled
```

```python
x_scaled.shape 
```

```python
# apply pca
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(x_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
# Agglomerative clustering
# to construct a dendogram

import scipy.cluster.hierarchy as sc

# plot the dendogram
plt.figure(figsize=(20,7))
plt.title("dendogram")


sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title("dendogram")
plt.xlabel("index")
plt.ylabel("Eucledian distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score

silhouette_coefficients=[]

for k in range(2,11):
    agglo= AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    agglo.fit(pca_scaled)
    silhouette_coefficients.append(silhouette_score(pca_scaled,agglo.labels_))

plt.plot(range(2,11),silhouette_coefficients)
plt.show()
```

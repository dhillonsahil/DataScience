```python
# divide dataset into dependent and dependent features
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

# Model Training

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

we dont scale output 

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled,y_train)
```

```python
## slopes of 8 features
model.coef_
```

these are 8 slope , 1 per feature

```python
model.intercept_
```

```python
## Prediction

y_pred_test = model.predict(X_test_scaled)
```

```python
y_pred_test
```

## Performace metrices cost function"

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(mean_squared_error(y_test,y_pred_test))
print(mean_absolute_error(y_test,y_pred_test))
print(np.sqrt(mean_squared_error(y_test,y_pred_test)))
```

```python
## Rsquare and adjusted r square
score = r2_score(y_test,y_pred_test)
print(score)
```

```python
## adjusted r square
1- (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

# Pickling

 Pickling in Python refers to the serialization of Python objects into a byte stream, which can be stored in a file or transmitted over a network. Unpickling is the inverse process, reconstructing the original Python object from the byte stream. The pickle module provides the necessary tools for these operations. 


```python
import pickle
pickle.dump(scaler,open('scaler.pkl','wb'))
pickle.dump(model,open('regressor.pkl','wb'))
```

```python
model_regressor =pickle.load(open('regressor.pkl','rb'))
```

```python
model_regressor.predict(X_test_scaled)
```

```python
standard_scaler= pickle.load(open('scaler.pkl','rb'))
```

```python
model_regressor.predict(standard_scaler.transform(X_test))
```

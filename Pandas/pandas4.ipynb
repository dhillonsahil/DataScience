````python
data2 = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C','A'],
    'Type': ['X', 'Y', 'X', 'Y', 'X', 'Y','X'],
    'Values': [10, 20, 30, 40, 50, 60,55]
}

df6 = pd.DataFrame(data2)

# Group by 'Category' and 'Type'
grouped2 = df6.groupby(['Category', 'Type'])

print(grouped2)
````

````python
# iterating over Groups
for name, group in df6.groupby('Category'):
    print(f"Group: {name}")
    print(group)
    print("\n")
````

````python
# Keep only groups where the sum of 'Values' is greater than 50
filtered = df6.groupby('Category').filter(lambda x: x['Values'].sum() > 50)
print(filtered)
````

````python
grouped3 = df6.groupby('Category')['Values'].agg(['sum', 'mean', 'max']) # The .agg() method allows you to apply multiple functions at once.
print(grouped3)
````

````python
# You can also use custom functions inside agg():

grouped4 = df6.groupby('Category')['Values'].agg(['sum', lambda x: x.max() - x.min()])
print(grouped4)
````

````python
 # Grouping with Missing Data By default, groupby() ignores NaN values. You can use dropna=False to include them.

df6.loc[2, 'Category'] = None  # Introduce NaN
grouped5 = df6.groupby('Category', dropna=False).sum()
print(grouped5)
````

````python
# Sorting Groups

#You can sort groups using .sort_values().

grouped6 = df6.groupby('Category')['Values'].sum().reset_index()
grouped_sorted = grouped6.sort_values(by='Values', ascending=False)
print(grouped_sorted)
````

````python
data_1 = pd.DataFrame({
    "key1":[1,2,4,5,6],
    "key2":[4,5,6,7,8],
    'key3':[3,4,5,6,6]
})
data_2 = pd.DataFrame({
    "key1":[1,2,45,78,6],
    "key4":[96,5,6,78,8],
    'key5':[3,4,5,52,6]
})
````

````python
data_1
````

````python
data_2
````

````python
pd.merge(data_1,data_2) # how param by default inr , so it uses inner key intersection where same value is present on - common value is key 1 so key 1 intersection is done
# how="left" or  "right" or "outer" or 'cross'
````

LEFT JOIN (how="left")

1. Keeps all rows from the left DataFrame (df1).
1. If a matching row is found in df2, it merges the values.
1. If no match, it fills with NaN.


RIGHT JOIN (how="right")

1. Keeps all rows from the right DataFrame (df2).
2. If a matching row is found in df1, it merges the values.
3. If no match, it fills with NaN.

 OUTER JOIN (how="outer")

1. Keeps all rows from both DataFrames.
2. If a row exists in only one DataFrame, missing values are NaN.

Cross Join (Cartesian Product)

1. Combine all rows from both DataFrames without any matching condition. ( will work if no same key in both too)

````python
pd.merge(data_1,data_2,how='left')
````

````python
pd.merge(data_1,data_2,how='right')
````

````python
pd.merge(data_1, data_2, how="outer")
````

````python
data_3 = pd.DataFrame({
    "key1":[1,2,4,5,6],
    "key2":[4,5,6,7,8],
    'key3':[3,4,5,6,6]
})
data_4 = pd.DataFrame({
    "key6":[1,2,45,78,6],
    "key4":[96,5,6,78,8],
    'key5':[3,4,5,52,6]
})
````

````python
pd.merge(data_3,data_4,how='cross')
````

This performs a CROSS JOIN, also known as a Cartesian Product. Let’s break it down.
🔥 What is a Cross Join?

A cross join merges every row from the first DataFrame with every row from the second DataFrame, regardless of whether they have a common column.

Formula for total rows in cross join:
Total Rows=(Rows in data_3)×(Rows in data_4)
Total Rows=(Rows in data_3)×(Rows in data_4)
📌 Your Data:

    data_3 has 5 rows
    data_4 has 5 rows
    So, total rows in the output = 5 × 5 = 25 rows

✅ Understanding the Output

Each row from data_3 is paired with every row from data_4.

Example:
If data_3 had this row:

key1  key2  key3
1     4     3

It gets repeated for every row in data_4, like:

| key1 | key2 | key3 | key6 | key4 | key5 |
|------|------|------|------|------|------|
| 1    | 4    | 3    | 1    | 96   | 3    |
| 1    | 4    | 3    | 2    | 5    | 4    |
| 1    | 4    | 3    | 45   | 6    | 5    |
| 1    | 4    | 3    | 78   | 78   | 52   |
| 1    | 4    | 3    | 6    | 8    | 6    |

Then the next row in data_3 (e.g., key1=2, key2=5, key3=4) will also be paired with all 5 rows from data_4.

This repeats for all 5 rows of data_3, leading to 25 total rows.

#### Understanding `left_on`, `right_on`, `how` in Pandas `merge()`

When merging two DataFrames in Pandas, you can specify which columns should be used as keys using **`left_on`** and **`right_on`**. You also specify the type of join using the **`how`** parameter.

#### 1️⃣ `left_on` and `right_on`

- **`left_on`** → Specifies the column in the **left DataFrame** to use as the key for merging.  
- **`right_on`** → Specifies the column in the **right DataFrame** to use as the key for merging.  

### #🔹 Example:

```python
import pandas as pd

df1 = pd.DataFrame({
    "A": [1, 2, 3],
    "B": ["X", "Y", "Z"]
})

df2 = pd.DataFrame({
    "C": [1, 2, 4],
    "D": ["P", "Q", "R"]
})

merged_df = pd.merge(df1, df2, left_on="A", right_on="C", how="inner")
print(merged_df)
```

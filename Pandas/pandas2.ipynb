The main difference between .loc and .iloc in Pandas is that .loc is used for label-based indexing, allowing you to select rows and columns by their names, while .iloc is used for integer-based indexing, allowing you to select rows and columns by their numerical positions. This means .loc includes the endpoint in a range, whereas .iloc does not.

```python
df.loc[0:2,['PassengerId','Fare']] # get 1st 2 rows from Columns Passenger Id and Fare
```

```python
df.loc[0:2,['PassengerId','Fare']] # 1st param is always row and 2nd is Fare
```

```python
df.iloc[0:2,[0,1]] # iloc requires the index for columns too
```
